```python
import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/Реализация всех алгоритмов/data_banknote_authentication.txt', header=None)
df.columns = ['variance', 'skewness', 'curtosis', 'entropy', 'target']
X, y = df.iloc[:,:4], df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
class MyKNNClf():
    def __init__(self, k=3,metric='euclidean'):
        self.k = k
        self.X = 0
        self.y = 0
        self.train_size = 0
        self.metric = metric
    def __str__(self):
        attributes = ', '.join(f"{key}={value}" for key, value in vars(self).items())
        return f"MyKNNClf class: {attributes}"
    def __repr__(self):
        attributes = ', '.join(f"{key}={value}" for key, value in vars(self).items())
        return f"MyKNNClf class: {attributes}"

    def fit(self, X_fit: pd.DataFrame, y_fit: pd.Series):
      self.X = X_fit.copy()
      self.y = y_fit.copy()
      self.train_size = (self.X.shape[0],self.X.shape[1])

    def metrics(self,X_test: pd.DataFrame):
      if self.metric == 'euclidean':
        sum = np.zeros(self.X.shape[0])
        for b in range(len(self.X.columns)):
          sum+=(self.X.loc[:,self.X.columns[b]] - X_test[self.X.columns[b]])**2
        res = np.sqrt(sum)
        return res
      if self.metric == 'chebyshev':
        list_values =[]
        sum = np.zeros(self.X.shape[0])
        for b in range(len(self.X.columns)):
          list_values.append(abs(self.X.loc[:,self.X.columns[b]] - X_test[self.X.columns[b]]))
        res=[]
        for i in range(X_train.shape[0]):
            max_value_i = max([s.iloc[i] for s in list_values])
            res.append(max_value_i)
        return pd.Series(res,index=X_train.index)

    def evristic(self, X_test: pd.DataFrame,chanse):
      index= self.metrics(X_test).sort_values()[:self.k].index
      if chanse ==0:
        if np.count_nonzero(self.y.loc[index] == 1) == np.count_nonzero(self.y.loc[index] == 0):
          return(1)
        return(self.y.loc[index].mode().values[0])
      if chanse ==1:
        one_count = np.count_nonzero(self.y.loc[index] == 1)
        return(one_count/self.k)

    def predict(self, X_test: pd.DataFrame):
        y_pred_vec = X_test.apply(lambda x: self.evristic(x,0),axis=1)
        return y_pred_vec

    def predict_proba(self, X_test: pd.DataFrame):
        y_pred_vec = X_test.apply(lambda x: self.evristic(x,1),axis=1)
        return y_pred_vec
```
